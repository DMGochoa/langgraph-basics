```python
# Estos son algunos de los mensajes que se pueden usar en el agente simple. Puedes editarlos o agregar nuevos mensajes según sea necesario.
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage, ToolMessage

ai_message = AIMessage(content="Hola, soy un agente simple. ¿En qué puedo ayudarte?")
```

```python
print(ai_message.text)
```

```python
human_message = HumanMessage(content="¿Cuál es tu nombre?")
print(human_message.text)
```

```python
history = [ai_message, human_message]
for message in history:
    print(f"{message.__class__.__name__}: {message.text}")
```

```python
history += [AIMessage(content="Mi nombre es SimpleAgent.")]
for message in history:
    print(f"{message.__class__.__name__}: {message.text}")
```
