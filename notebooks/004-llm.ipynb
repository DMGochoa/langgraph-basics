```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

qwen = ChatOpenAI(
    base_url="http://192.168.1.2:1234/v1",
    #api_key="lm-studio",  # LMStudio no requiere clave real, pero el campo es obligatorio
    model="qwen/qwen3-4b-thinking-2507",
)
response = qwen.invoke("Hello, how are you?")
```

```python
response.content
```

```python
chatgpt = ChatOpenAI(
    base_url="http://192.168.1.2:1234/v1",
    #api_key="lm-studio",  # LMStudio no requiere clave real, pero el campo es obligatorio
    model="openai/gpt-oss-20b",
)
response_chatgpt = chatgpt.invoke("Hello, how are you?")
```

```python
response_chatgpt.content
```

```python
from langchain_core.messages import AIMessage, BaseMessage, HumanMessage, SystemMessage

msg_1 = SystemMessage(content="You are a helpful assistant that talks in spanish and responds like a pirate.")
msg_2 = HumanMessage(content="Hola, me llamo Diego, ¿cómo estás?")
msg_3 = AIMessage(content="¡Ahoy, Diego! Estoy bien, gracias por preguntar. ¿En qué aventura nos embarcamos hoy?")
msg_4 = HumanMessage(content="¿Cuál es la capital de Francia?")

history = [msg_1, msg_2, msg_3, msg_4]
```

```python
response = qwen.invoke(history)
response.content
```

```python
response = chatgpt.invoke(history)
response.content
```

```python
from langchain_ollama import ChatOllama

gemma4 = ChatOllama(
        model="gemma4:e4b",
        base_url="http://192.168.1.2:11434"
    )
response = gemma4.invoke(history)
response.content
```

```python
from langchain.chat_models import init_chat_model

qwen = init_chat_model(
    model="openai:qwen/qwen3-4b-thinking-2507",
    base_url="http://192.168.1.2:1234/v1"
)
response = qwen.invoke(history)
response.content
```
