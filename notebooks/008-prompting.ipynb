```python
from langchain_core.prompts import PromptTemplate
```

```python
template = """\
    # Instrucciones para el asistente de IA

    Recuerda que hoy es '{fecha_actual}'

    ## Rol
    Eres un experto en marketing digital con 10 años de experiencia.

    ## Tarea
    Analiza el siguiente texto de un anuncio y proporciona 3 sugerencias para mejorarlo, enfocandote en el llamado a la accion.

    ## Texto del anuncio
    '{texto_anuncio}'

    ## Formato de salida
    - *Sugerencia 1:* [Tu primera sugerencia]
    - *Sugerencia 2:* [Tu segunda sugerencia]
    - *Sugerencia 3:* [Tu tercera sugerencia]
"""
prompt = PromptTemplate.from_template(template)
```

```python
print(
    prompt.format(
        fecha_actual= "2026-04-25",
        texto_anuncio="Este es un anuncio de prueba"
    )
)
```

```python
# si queremos llenar algun valor por defecto o que se autorellene podemos hacer lo siguiente:
from datetime import datetime
today = datetime.now().strftime("%Y-%m-%d")
prompt = PromptTemplate.from_template(template, partial_variables={"fecha_actual":today})
```

```python
print(
    prompt.format(
        texto_anuncio="Este es un anuncio de prueba"
    )
)
```
