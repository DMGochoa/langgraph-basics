```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

qwen = init_chat_model(
    model="openai:qwen/qwen3-4b-2507",
    base_url="http://100.105.94.89:1234/v1",
    profile={
        "structure_output": True
    }
)
response = qwen.invoke("Hola soy diego y mi numero es +57 312 034 1234")
print(response.content)
```

```python
# Aqui ya que estamos agregando informacion debemos de usar el LLM para poder organizar la informacion
from pydantic import BaseModel, Field

class ContactInfo(BaseModel):
    name: str = Field(
        description="The name of the person"
    )
    email: str = Field(
        description="The email address of the person"
    )
    phone: str = Field(
        description="The phone number of the person"
    )
    tone: int = Field(
        description="The tone of the conversation",
        ge=0,
        le=100
    )
    sentiment: str = Field(
        description="The sentiment of the conversation"
    )
    age: int =Field(
        description="The age of the person"
    )

llm_with_structured_output = qwen.with_structured_output(schema=ContactInfo)
```

```python
messages = [
    ("system", "You are a helpfull assistant that can extract contact information from a given text."),
    ("user", "Hola soy diego y mi numero es +57 312 034 1234"),
]
response = llm_with_structured_output.invoke(
    messages
)
print(response)
```
